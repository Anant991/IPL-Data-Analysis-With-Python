# Bat First Winning Records

```python
# Extracting the records when a team won batting first

batting_first=ipl[ipl["win_by_runs"]!=0]
```

```python
batting_first
```

```python
batting_first_1=ipl[ipl["win_by_wickets"]==0]
```

```python
batting_first_1
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first["win_by_runs"])
plt.title("First Batting Team Won by Runs")
plt.xlabel("Runs")
plt.show()
```

```python
# Finding the number of wins of each team after batting first

batting_first["winner"].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.bar(list(batting_first["winner"].value_counts()[0:5].keys()),list(batting_first["winner"].value_counts()[0:5]),color=["yellow","blue","green","pink","red"])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first["winner"].value_counts()),labels=list(batting_first["winner"].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
